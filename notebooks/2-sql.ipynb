````python
### Initialiserung einer (In-Memory)-Datenbank mit den bekannten Daten

import os
import pandas as pd
import sqlite3
from urllib.request import urlopen

GITHUB_URL = "https://raw.githubusercontent.com/markusschanta/w3schools-database/refs/heads/master/w3schools.sqllite.sql"
sql = urlopen(GITHUB_URL).read().decode()

db = sqlite3.connect("file::memory:?cache=shared")

db.executescript(sql)
db.commit()
````

````python
SQL = "SELECT * FROM customers"

pd.read_sql_query(SQL, db)
````

# WHERE

````python
SQL = """
SELECT ProductName, Price
FROM Products
WHERE Price > 50
"""

pd.read_sql_query(SQL, db)
````

````python
SQL = """
SELECT ProductName, Price
FROM Products
WHERE Price BETWEEN 10 AND 13
"""

pd.read_sql_query(SQL, db)
````

````python
SQL = """
SELECT *
FROM Customers
WHERE Country = 'germany'
"""

pd.read_sql_query(SQL, db)
````

````python
SQL = """
SELECT *
FROM Customers
WHERE Country LIKE '%land'
"""

pd.read_sql_query(SQL, db)
````

````python
SQL = """
SELECT *
FROM Customers
WHERE PostalCode LIKE '____'
"""

pd.read_sql_query(SQL, db)
````

# ORDER BY

# DISTINCT

````python
SQL = """
SELECT Country
FROM Customers
"""

pd.read_sql_query(SQL, db)
````

````python
SQL = """
SELECT DISTINCT Country
FROM Customers
"""

pd.read_sql_query(SQL, db)
````

````python
SQL = """
SELECT CategoryName, AVG(Price) as PriceAvg
FROM Products
LEFT JOIN Categories
  ON Products.CategoryID = Categories.CategoryID
GROUP BY CategoryName
ORDER BY PriceAvg DESC
"""

pd.read_sql_query(SQL, db)
````

# GROUP BY

````python
SQL = """
SELECT CategoryID, COUNT(ProductID)
FROM Products
GROUP BY CategoryID
"""

pd.read_sql_query(SQL, db)
````

````python
SQL = """
SELECT CategoryName, AVG(Price)
FROM Products
LEFT JOIN Categories
  ON Products.CategoryID = Categories.CategoryID
GROUP BY CategoryName
"""

pd.read_sql_query(SQL, db)
````

# JOIN

````python
SQL = """
SELECT *
FROM Products
LEFT JOIN Categories
  ON Products.CategoryID = Categories.CategoryID
"""

pd.read_sql_query(SQL, db)
````

````python
SQL = """
SELECT *
FROM Products
LEFT JOIN Categories
  ON Products.CategoryID = Categories.CategoryID
WHERE CategoryName='Produce'
"""

pd.read_sql_query(SQL, db)
````

````python
SQL = """
SELECT CategoryName, COUNT(ProductID) as ProductIDCount
FROM Products
LEFT JOIN Categories
  ON Products.CategoryID = Categories.CategoryID
GROUP BY CategoryName
"""

pd.read_sql_query(SQL, db)
````

````python
SQL = """
SELECT *
FROM Products p
LEFT JOIN Categories c
  ON p.CategoryID=c.CategoryID
LEFT JOIN Suppliers s
  ON p.SupplierID=s.SupplierID
"""

pd.read_sql_query(SQL, db)
````

````python
SQL = """SELECT s.SupplierName, AVG(p.Price) as PriceAvg
FROM Products p
LEFT JOIN Suppliers s
  ON p.SupplierID=s.SupplierID
GROUP BY s.SupplierName
ORDER BY PriceAvg DESC"""

pd.read_sql_query(SQL, db)
````
