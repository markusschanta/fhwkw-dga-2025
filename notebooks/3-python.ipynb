Einkommensteuer:

* bis 13.308 Euro:	0%
* bis 21.617 Euro:	20%
* bis 35.836 Euro:	30%
* bis 69.166 Euro:	40%
* bis 103.072 Euro: 48%
* bis 1.000.000 Euro: 50%
* ab 1.000.000 Euro: 55%

```python
def einkommenssteuer(bruttogehalt):
    steuer = 0
    restgehalt = bruttogehalt

    mehrbetrag = max(restgehalt - 1000000, 0)
    steuer = steuer + mehrbetrag * 0.55
    restgehalt = restgehalt - mehrbetrag

    mehrbetrag = max(restgehalt - 103072, 0)
    steuer = steuer + mehrbetrag * 0.5
    restgehalt = restgehalt - mehrbetrag

    mehrbetrag = max(restgehalt - 69166, 0)
    steuer = steuer + mehrbetrag * 0.48
    restgehalt = restgehalt - mehrbetrag

    mehrbetrag = max(restgehalt - 35836, 0)
    steuer = steuer + mehrbetrag * 0.4
    restgehalt = restgehalt - mehrbetrag

    mehrbetrag = max(restgehalt - 21617, 0)
    steuer = steuer + mehrbetrag * 0.3
    restgehalt = restgehalt - mehrbetrag

    mehrbetrag = max(restgehalt - 13308, 0)
    steuer = steuer + mehrbetrag * 0.2
    restgehalt = restgehalt - mehrbetrag

    return steuer
```

```python
einkommenssteuer(75000)
```
