Funktionen dienen in Python dazu, Code zu strukturieren und ihn wiederverwendbar zu machen. Eine einmal definierte Funktion kann an beliebigen, mehreren Stellen aufgerufen werden, um so komplexere Programmabl√§ufe aufzubauen.

## Einfache Funktionen

In Python k√∂nnen Funtionen wie folgt definiert und aufgerufen werden:

```python
# Definition der Funktion

def trennlinie_drucken():
    print("--------------------------")
```

Die Funktionsdefinition beginnt mit dem Schl√ºsselwort `def`, gefolgt vom Namen der Funktion (in diesem Beispiel `trennlinie_drucken`). Dem Namen folgt ein Klammerausdruck, in dem die Parameter (=Eingabewerte) der Funktion angegeben sind (hier keine; leere Klammer `()`). Nach dem Doppelpunkt folgen einger√ºckt die Befehle, die beim Aufruf der Funktion ausgef√ºhrt werden sollen (in diesem Beispiel ein `print`-Befehl).

Die so definierte Funktion `trennlinie_drucken` kann nun im restlichen Programm wie folgt aufgerufen werden:

```python
# Aufruf der Funktion

trennlinie_drucken()
print("RECHNUNG - DEIN SHOP")
trennlinie_drucken()
```

Der Aufruf der Funktion erfolgt √ºber den Namen, gefolgt von einem Klammerausdruck, in dem die Parameter (hier keine) √ºbergeben werden. So kann die Funktionaufgerufen werden, womit der K√∂rper der Funktion ausgef√ºhrt wird.

## Funktionen mit Parametern

Funktionen k√∂nnen, wie bereits angedeutet auch Eingabewerte √ºbergeben bekommen. Diese stehen bei der Ausf√ºhrung dann innerhalb der Funktion zur Verf√ºgung und k√∂nnen den Programmablauf beeinflussen:

```python
# Definition der Funktion

def artikel_drucken(anzahl, name, preis):
    ausgabe = str(anzahl) + " x " + name
    ausgabe = ausgabe + " (" + str(preis) + " EUR pro St√ºck): "
    ausgabe = ausgabe + str(anzahl * preis) + " EUR"
    print(ausgabe)
```

In der Definition der Funktion `artikel_drucken` sind drei Parameter (`anzahl`, `name`, `preis`) angegeben. Im K√∂rper der Funktion werden diese drei Parameter verwendet, um die Eingaben entsprechend zu verarbeiten (hier ausdrucken).

Die Funktion `artikel_drucken` kann nun im weiteren wie folgt verwendet werden:

```python
# Aufruf der Funktion

artikel_drucken(3, "Apfel", 1.50)
```

Damit k√∂nnen wir das Beispiel nun wie folgt erweitern:

```python
trennlinie_drucken()
print("RECHNUNG - DEIN SHOP")
trennlinie_drucken()

artikel_drucken(3, "Apfel", 1.50)
artikel_drucken(2, "Banane", 0.80)
artikel_drucken(1, "Kirsche", 2.00)
trennlinie_drucken()
```

## Funktionen mit R√ºckgabewerten

Neben Eingabewerten (Parametern) k√∂nnen Funktionen in Python auch R√ºckgabewerte zur√ºckliefern. Eine Funktion mit R√ºckgabewerten kann beispielsweise wie folgt defininert werden:

```python
def berechne_gesamtpreis(artikelliste):
    gesamtpreis = 0
    for artikel in artikelliste:
        menge = artikel[0]
        preis = artikel[2]
        gesamtpreis = gesamtpreis + (menge * preis)
    return gesamtpreis
```

Die Funtkion `berechne_gesamtpreis` wird wie andere Funktionen beginnend mit dem Schl√ºsselwort `def`, gefolgt vom Namen der Funktion (`berechne_gesamtpreis`) und den Parametern (`artikelliste`) definiert. Nach dem Doppelpunkt folgen die Anweisungen, die beim Aufruf der Funktion ausgef√ºhrt werden.

Der R√ºckgabewert der Funktion wir mit dem Schl√ºsselwort `return` defininert.

Die Funktion kann nun wie folgt verwendet werden:

```python
artikelliste = [
    (3, "Apfel", 1.50),
    (2, "Banane", 0.80),
    (1, "Kirsche", 2.00)
]

print(berechne_gesamtpreis(artikelliste))
```

Aus den bisher definierten Funktionen k√∂nnen wir nun beispielsweise folgenden Programmablauf zusammensetzen:

```python
artikelliste = [
    (3, "Apfel", 1.50),
    (2, "Banane", 0.80),
    (1, "Kirsche", 2.00)
]

trennlinie_drucken()
print("RECHNUNG - DEIN SHOP")
trennlinie_drucken()

for artikel in artikelliste:
    artikel_drucken(artikel[0], artikel[1], artikel[2])
trennlinie_drucken()

print("Gesamtpreis: " + str(berechne_gesamtpreis(artikelliste)) + " EUR")
trennlinie_drucken()
```
